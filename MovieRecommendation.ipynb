```python
#using cosine similarity to calculate similarity row-wise
item_similarity = cosine_similarity(df_std.T) #taking the transpose
print(item_similarity)
```

```python
#convertinf into dataframe from the above numoy array
df = pd.DataFrame(item_similarity,index =ratings.columns,columns = ratings.columns )
df
```

```python
#making recommendations

def get_similar_movies(name,rating):
    similar_score = df[name]*rating
    similar_score = similar_score.sort_values(ascending=False)
    return similar_score
print(get_similar_movies("Movie5",4))
```
